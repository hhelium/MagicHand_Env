```python
import pandas as pd
import numpy as np
import csv
import os
import copy
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("monitor.csv")
obj = pd.read_csv("poPdAb2_list.csv")
obj = obj.iloc[:4,:]
l = list(obj["Object"].unique())
```

```python
for i in range(4):
    o = data[data["i"]==i]
    s = o[o["s"]==1]
    print('The success rate of {} is {} out of {} = {}'.format(l[i], len(s), len(o), len(s)/len(o)))
```

```python
ed = np.load("episodeData.npy", allow_pickle = True)
ed = ed.tolist()[0]
nd = copy.deepcopy(ed)
for i in range(len(nd)):
    nd[i].pop(-1)
```

```python
# epicode reward 
er = []
#for i in range(len(nd)):
for i in range(100):
    er.append(sum(nd[i]))

plt.plot(er)
plt.grid()
plt.xlabel('Episodes')
plt.ylabel('Reward')
plt.show()
```

```python
aer = []
#for i in range(len(nd)):
for i in range(300):

    aer.append(sum(nd[i])/len(nd[i]))
plt.plot(aer)
plt.grid()
plt.xlabel('Episodes')
plt.ylabel('Average Reward')
plt.show()              
```
